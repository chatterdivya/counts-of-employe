```python
import pandas as pd
import matplotlib.pyplot as plt

# Load a sample CSV file (you can replace with your own)
data = {
    "Name": ["Alice", "Bob", "Charlie", "Diana"],
    "Age": [25, 30, 35, 28],
    "Salary": [50000, 60000, 55000, 62000],
    "Department": ["HR", "IT", "Finance", "IT"]
}
df = pd.DataFrame(data)

# --- Calculate average of a selected column ---
column_for_mean = "Salary"

if column_for_mean in df.columns:
    mean_val = df[column_for_mean].dropna().mean()
    print(f"\nAverage of '{column_for_mean}': {mean_val:.2f}")
else:
    print(f"\nColumn '{column_for_mean}' not found in data.")

# --- Create a bar chart for a categorical column ---
cat_col = "Department"
counts = df[cat_col].value_counts()

plt.figure(figsize=(6,4))
counts.plot(kind='bar', color='skyblue')
plt.title(f"Count of Employees by {cat_col}")
plt.xlabel(cat_col)
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```
